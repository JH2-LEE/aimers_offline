#### 가격포지셔닝 컬럼

- 가격포지셔닝 - 같은 대분류 안에서 평균가격바탕으로 표준화

```python
train_data['대분류'].value_counts()
```

- 판매금액 데이터 만들기

```python
info = train_data.iloc[:, :7]
train = train_data.iloc[:, 7:]
sales = sales_data.iloc[:, 7:]
price = sales.div(train)
price1 = pd.concat([info, price], axis=1)
```

```python
price1
```

```python
def mean_excluding_zeros(row):
    non_zero_values = row[row != 0]
    if len(non_zero_values) > 0:
        return non_zero_values.mean()
    else:
        return 0  # Return 0 if all values are zeros

price_means = price.apply(mean_excluding_zeros, axis=1)
```

```python
price_means1 = pd.concat([train_data[['대분류']], price_means],axis=1)
price_means1.columns = ['대분류', '가격평균']
price_means1
```

```python
price_means1[price_means1['가격평균'].isna()]
```

```python
price_means1['가격평균'].fillna(0, inplace=True)
```

- 대분류별 그룹화

```python
brand_stats = price_means1.groupby('대분류')['가격평균'].agg(['mean', 'std']).reset_index()
brand_stats.rename(columns={'mean': 'MeanPrice', 'std': 'StdDevPrice'}, inplace=True)
```

```python
price_means2 = pd.merge(price_means1, brand_stats, on='대분류', how='left')
```

```python
price_means2
```

```python
price_means2['ZScore_대분류'] = (price_means2['가격평균'] - price_means2['MeanPrice']) / price_means2['StdDevPrice']
```

```python
price_means2
```

```python
merged_df5 = pd.concat([merged_df4, price_means2[['가격평균', 'ZScore_대분류']]], axis=1)
merged_df5[merged_df5['브랜드']=='B002-00002']
```
