```python
import pandas as pd
import numpy as np

train = pd.read_csv('/root/workspace/open/train.csv')

pd.set_option('display.max_columns', None)
```

```python
shop_df = pd.DataFrame(train['쇼핑몰'].value_counts()).reset_index().reset_index()
```

```python
shop_df.columns = ['쇼핑몰순위', '쇼핑몰',' 쇼핑몰개수']
```

```python
shop_df
```

```python
new = train.merge(shop_df, left_on='쇼핑몰', right_on='쇼핑몰', how='left')
```

```python
new.head()
```

```python
new_df = new[['ID', '쇼핑몰순위',' 쇼핑몰개수']]
```

```python
features = pd.read_csv("/root/workspace/data/features.csv", index_col = 0)
```

```python
features2 = features.merge(new_df, left_on='ID', right_on='ID', how='left')
```

```python
features2.to_csv('/root/workspace/data/features2.csv')
```
